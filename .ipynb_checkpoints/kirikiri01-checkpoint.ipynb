```python
import cv2
```

```python
def generate_dataset(id):
    face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

    def face_cropped(img):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, 1.3, 5)
        cropped_faces = []
        for (x, y, w, h) in faces:
            cropped_face = img[y:y+h, x:x+w]
            cropped_faces.append(cropped_face)
        return cropped_faces

    cap = cv2.VideoCapture(0)  # Use the default camera index, change to 1 if needed
    if not cap.isOpened():
        print("Error: Failed to open camera.")
        return

    
    img_id = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Failed to capture frame.")
            break

        detected_faces = face_cropped(frame)
        if detected_faces:
            for face in detected_faces:
                img_id += 1
                
                face_resized = cv2.resize(face, (200, 200))
                face_gray = cv2.cvtColor(face_resized, cv2.COLOR_BGR2GRAY)
                file_name_path = f"data/user.{id}.{img_id}.jpg"
                cv2.imwrite(file_name_path, face_gray)
                cv2.putText(face_resized, str(img_id), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
                cv2.imshow("Cropped face", face_resized)

        if cv2.waitKey(1) == 13 or int(img_id) == 200:
            break

    cap.release()
    cv2.destroyAllWindows()
    print("Collecting samples is completed.")

id=1
while(True):
    x=int(input("For capturing enter 1 otherwise any number"))
    if x==1:
        generate_dataset(id)
        id+=1
    else:
        break

```

```python
import os
import cv2
from PIL import Image #pip install pillow
import numpy as np    # pip install numpy
 
def train_classifier(data_dir):
    path = [os.path.join(data_dir, f) for f in os.listdir(data_dir)]
     
    faces = []
    ids = []
     
    for image in path:
        img = Image.open(image).convert('L')
        imageNp = np.array(img, 'uint8')
        id = int(os.path.split(image)[1].split(".")[1])
         
        faces.append(imageNp)
        ids.append(id)
         
    ids = np.array(ids)
     
    # Train and save classifier
    clf=cv2.face.LBPHFaceRecognizer_create(); 
    clf.train(faces,ids)
    clf.write("classifier.xml")
train_classifier("data")
```

```python
import cv2
import numpy as np

def draw_boundary(img, classifier, scaleFactor, minNeighbors, color, text, clf):
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    features = classifier.detectMultiScale(gray_img, scaleFactor, minNeighbors)
    
    for (x, y, w, h) in features:
        cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
        
        id, pred = clf.predict(gray_img[y:y+h, x:x+w])
        confidence = int(100 * (1 - pred / 300))
        
        if confidence > 70:
            if id == 1:
                cv2.putText(img, "Indra", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
            elif id == 2:
                cv2.putText(img, "Indra2", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
            # Add more elif blocks for other IDs as needed
        else:
            cv2.putText(img, "UNKNOWN", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 1, cv2.LINE_AA)
            # You might want to display confidence level here as well

    return img

# Load classifier
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
clf = cv2.face.LBPHFaceRecognizer_create()
clf.read("classifier.xml")

video_capture = cv2.VideoCapture(0)  # Use default camera index

if not video_capture.isOpened():
    print("Error: Failed to open camera.")
    exit()

while True:
    ret, img = video_capture.read()
    
    if not ret:
        print("Error: Failed to capture frame.")
        break
    
    img = draw_boundary(img, faceCascade, 1.3, 6, (255, 255, 255), "Face", clf)
    cv2.imshow("Face Detection", img)
    
    if cv2.waitKey(1) == 13:  # Press Enter to exit
        break

video_capture.release()
cv2.destroyAllWindows()
```
